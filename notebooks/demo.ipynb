# Speech Emotion Recognition - Demo

This notebook demonstrates the end-to-end inference and SHAP explainability pipeline for our CNN-LSTM emotion recognition system.

```python
import os
import sys
sys.path.append('..')

import yaml
import torch
from src.pipeline.data_pipeline import get_dataloaders
from src.models.cnn_lstm import CNNLSTM
from src.explainability.shap_explainer import ModelExplainer
from src.explainability.visualizations import plot_shap_spectrogram

# Load configuration
with open('../configs/config.yaml', 'r') as f:
    config = yaml.safe_load(f)
```

```python
# Setup Data and Model
train_loader, val_loader, test_loader = get_dataloaders('non_existent.csv', config)
model = CNNLSTM(config).eval()
```

```python
# Fetch a sample and generate explanation
bg_data, _ = next(iter(train_loader))
test_data, labels = next(iter(test_loader))

explainer = ModelExplainer(model, bg_data[:5], device='cpu')
sample = test_data[:1]
target = labels[0].item()

shap_values = explainer.explain(sample, target_class=target)
plot_shap_spectrogram(sample[0, 0].numpy(), shap_values[0, 0])
```
